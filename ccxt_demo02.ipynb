```python
import ccxt
#支持的交易所
ccxt.exchanges
```

```python
# 创建一个交易所对象
exchange = ccxt.gate({
    "apiKey": '',
    "secret":'',
    'enableRateLimit': True
})
```

```python
#列出对象的所有方法 dir(exchange)
print(exchange.id)
print(exchange.name)
print(exchange.has['publicAPI'])
print(exchange.has['privateAPI'])
print(exchange.timeframes)
print(exchange.timeout/1000)
print(exchange.rateLimit)
print(exchange.iso8601(exchange.milliseconds()))
```

```python
marks = exchange.load_markets()
```

```python
# 定义一个交易对
symbol = 'WLD/USDT'
# 加载单个交易对信息
markbtc = exchange.market(symbol)
```

```python
#获取账户
balance = exchange.fetch_balance()
balance
```

```python
#获取盘口数据
ticker = exchange.fetch_ticker(symbol)
# exchange.fetchTickers(['ETH/USDT','BTC/USDT'])
#最新价
last = ticker['last']
last
```

```python
#获取盘口数据
order_book = exchange.fetch_order_book(symbol)
# print('bids',order_book['bids'])#卖
# print('asks',order_book['asks'])#买

bid = order_book['bids'][0][0]
ask = order_book['asks'][0][0]
spread = bid-ask
print('买价：{:.2f},卖价：{:.2f},价差：{:.2f}'.format(bid,ask,spread))
```
