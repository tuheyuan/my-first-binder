```python
import ccxt

def fetch_ohlcv():
    exchange = ccxt.poloniex()  # 初始化 poloniex 交易所
    
    symbol = 'BTC/USDT'  # 交易对
    timeframe = '1m'  # K线时间周期（例如 '1m'，'5m'，'1h' 等）
    since = None  # 获取最近的K线数据，如果需要从特定时间点开始获取，可以提供时间戳（以毫秒为单位）
    limit = 100  # 要获取的K线数据数量，默认可以设置为100

    try:
        ohlcv_data = exchange.fetch_ohlcv(symbol, timeframe, since, limit)
        # 显示获取的OHLCV数据
        for candle in ohlcv_data:
            print(candle)
    except ccxt.BaseError as e:
        print(f"An error occurred: {e}")

# 调用函数获取OHLCV数据
fetch_ohlcv()
```
